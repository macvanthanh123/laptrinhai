### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/titanic_train.csv)

### Step 3. Assign it to a variable titanic 

```python
df=pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/titanic_train.csv')
df
```

### Step 4. Set PassengerId as the index 

```python
df.set_index('PassengerId', inplace=True)
df
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
gender_counts = df['Sex'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Male/Female Proportion')
plt.axis('equal')

plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
age=df['Age']
fare=df['Fare']
sex=df['Sex'].map({'male':'b','female':'r'})
plt.scatter(age,fare,c=sex)
plt.xlabel('age')
plt.ylabel('fare')
plt.title('scatterplot with the Fare payed and the Age')
plt.show()
```

### Step 7. How many people survived?

```python
survived=df['Survived'].sum()
print(f"there are {survived} people")
```
