### Step 1. Import the necessary libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Step 2. Import the df2set from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/df2/pandas/tips.csv).

### Step 3. Assign it to a variable called tips

```python
tips=pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv')
tips
```

### Step 4. Delete the Unnamed 0 column

```python
df2 =tips.drop(columns='Unnamed: 0')
df2
```

### Step 5. Plot the total_bill column histogram

```python
total_bill=df2['total_bill']
plt.hist(total_bill)
plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
tip	= df2['tip']
plt.scatter(tip,total_bill)
plt.xlabel("tip")
plt.ylabel('total_bill')

plt.show()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
size= df2['size']
plt.scatter(tip,total_bill,label='total_bill')
plt.scatter(tip,size,label='size')
plt.title('relationship of total_bill, tip and size.')
plt.xlabel("tip")

plt.legend()
plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
day=df2['day']
plt.bar(day,total_bill)
plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
plt.scatter(df2['tip'], df2['day'], c=df2['sex'].map({'Male': 'blue', 'Female': 'red'}))
plt.xlabel('Tip')
plt.ylabel('Day')
plt.title('Scatter Plot of Tip vs Day by Sex')
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
plt.boxplot(
    [df2[df2['time'] == 'Lunch']['total_bill'],
     df2[df2['time'] == 'Dinner']['total_bill']],
    labels=['Lunch', 'Dinner']
)


plt.xlabel('Time')
plt.ylabel('Total Bill')
plt.title('Box Plot of Total Bill per Day by Time')


plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))


ax1.hist(df2[df2['time'] == 'Dinner']['tip'], bins=10, color='skyblue', edgecolor='black')
ax1.set_xlabel('Tip')
ax1.set_ylabel('Frequency')
ax1.set_title('Histogram of Tip - Dinner')

ax2.hist(df2[df2['time'] == 'Lunch']['tip'], bins=10, color='lightgreen', edgecolor='black')
ax2.set_xlabel('Tip')
ax2.set_ylabel('Frequency')
ax2.set_title('Histogram of Tip - Lunch')
plt.tight_layout()
plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
plt.figure(figsize=(10, 5))
plt.scatter(
    df2[(df2['sex'] == 'Male') & (df2['smoker'] == 'No')]['total_bill'],
    df2[(df2['sex'] == 'Male') & (df2['smoker'] == 'No')]['tip'],
    label='Male (No-smoker)',
    color='blue'
)

plt.scatter(
    df2[(df2['sex'] == 'Male') & (df2['smoker'] == 'Yes')]['total_bill'],
    df2[(df2['sex'] == 'Male') & (df2['smoker'] == 'Yes')]['tip'],
    label='Male (Smoker)',
    color='red'
)

plt.scatter(
    df2[(df2['sex'] == 'Female') & (df2['smoker'] == 'No')]['total_bill'],
    df2[(df2['sex'] == 'Female') & (df2['smoker'] == 'No')]['tip'],
    label='Female (No-smoker)',
    color='green'
)

plt.scatter(
    df2[(df2['sex'] == 'Female') & (df2['smoker'] == 'Yes')]['total_bill'],
    df2[(df2['sex'] == 'Female') & (df2['smoker'] == 'Yes')]['tip'],
    label='Female (Smoker)',
    color='orange'
)

plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter Plot of Total Bill and Tip by Gender and Smoking Status')
plt.legend()
plt.show()
```
