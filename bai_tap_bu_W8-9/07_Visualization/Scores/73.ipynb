### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Step 2. Create the DataFrame that should look like the one below.

```python
a={
    'first_name':['Jason','Molly','Tina','Jake','Amy'],
    'last_name':['Miller','Jacobson','Ali','Milner','Cooze'],
    'age':[42,54,36,24,73],
    'female':[0,1,1,0,1],
    'preTestScore':[4,24,31,2,3],
    'postTestScore':[25,94,37,62,70],
}
df=pd.DataFrame(a)
df
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
age=df['age']
preTestScore=df['preTestScore']
postTestScore=df['postTestScore']
plt.grid(ls='--')
plt.scatter(age,preTestScore,label='preTestScore')
plt.scatter(age,postTestScore,label='postTestScore')
plt.title('Scatterplot of preTestScore and postTestScore',fontsize=15)
plt.xlabel('age',fontsize=15)
plt.xticks(age)
plt.legend()
plt.show()
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
female=df['female']
plt.grid(ls='--')
plt.scatter(preTestScore,female)
plt.scatter(postTestScore,female)
plt.title('Scatterplot of preTestScore and postTestScore',fontsize=15)
plt.xlabel('age',fontsize=15)
plt.xticks()
plt.legend()
plt.show()
    
```
